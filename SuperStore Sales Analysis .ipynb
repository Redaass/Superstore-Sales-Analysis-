# Exploratory Data Analysis 

```python
# create new columns 
data['revenue'] = data ['sales']/data['quantity']
```

```python
data.head()
```

```python
data
```

# Region

```python
# ploting a bar chart for region vs revenue 
plt.figure (figsize = (20,10))
revenue_region = data.groupby(['region'], as_index = False )['revenue'].sum().sort_values(by='revenue' ,  ascending = False ).head(8)

sns.barplot (x = 'region' , y = 'revenue' , data = revenue_region)
```

From above we can see that the most region with highest revenue are Central , South and EMEA respectively. 

# Country 

```python
plt.figure (figsize = (20,4))
revenue_country = data.groupby (['country'], as_index = False )['revenue'].sum().sort_values (by='revenue' ,  ascending = False ).head(5)

sns.barplot( x = 'country' , y = 'revenue' , data = revenue_country)
```

From above we can see that the most country with high revenue are United States , Australia and France respectively. 

# State 

```python
plt.figure (figsize = (20,4))
revenue_country = data.groupby (['state'], as_index = False )['revenue'].sum().sort_values (by='revenue' ,  ascending = False ).head(10)

sns.barplot( x = 'state' , y = 'revenue' , data = revenue_country)
```

From above we can see that the most state with highest revenue are England , California  and ile-de-France respectively.
